```python
f = open("train_tweets_preprocessed.txt","r", encoding="utf8")
```

```python
t = open("test_tweets_preprocessed.txt","r", encoding="utf8")
```

```python
mySentenceList = []
myScoreList = []
```

```python
myTestSentenceList = []
myTestScoreList = []
```

```python
for line in f:
    line =line.strip()
    splittedLine = line.split(" ")
    score = splittedLine[len(splittedLine)-1]
    myScoreList.append((float(score.strip())+1)/2)
    del splittedLine[len(splittedLine)-1]
    sentence =" ".join(splittedLine)
    mySentenceList.append(sentence)
```

```python
for line in t:
    line =line.strip()
    splittedLine = line.split(" ")
    score = splittedLine[len(splittedLine)-1]
    print(score.strip())
    myTestScoreList.append((float(score.strip())+1)/2)
    del splittedLine[len(splittedLine)-1]
    sentence =" ".join(splittedLine)
    myTestSentenceList.append(sentence)
```
